```python
#Analyze Investment Avenues: Examine the column containing information about different 
#investment avenues (e.g., equity, mutual funds).
```

```python
all_investment=[ 'Mutual_Funds', 'Equity_Market',
       'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold',
      ]
```

```python
all_investment
```

```python
investment_counts = df[all_investment].apply(pd.to_numeric, errors='coerce').sum()
```

```python
investment_counts
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
a= investment_counts.plot(kind='bar',color='lightpink',edgecolor='black')
plt.title("Analysing Different Investment Avenue",fontsize=10, fontweight='bold')
plt.xlabel("all_investment")
plt.ylabel("Count")
plt.xticks(rotation=45)
# Show value on top of bar
for bar in a.patches:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{int(height)}', ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
investment_counts
```

```python
#Frequency Analysis: Determine the investment avenue with the highest frequency or occurrence.
```

```python
# Find most preferred investment
most_preferred = investment_counts.idxmax()   
frequency = investment_counts.max()         
```

```python
print("Most Preferred Investment Avenue:", most_preferred)
print("Frequency:", frequency)
```

```python
# Create a new Series to plot
most_preferred_series = pd.Series({most_preferred: frequency})
#visualize....
plt.figure(figsize=(4, 4))
bars = most_preferred_series.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title("Most Preferred Investment Avenue", fontsize=12)
plt.xlabel("Investment Avenue")
plt.ylabel("Count")
# Show value on top of bar
for bar in bars.patches:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{int(height)}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Rotate x-ticks if needed
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
